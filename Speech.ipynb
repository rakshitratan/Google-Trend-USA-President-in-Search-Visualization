```python
import requests # to request the standard html
from bs4 import BeautifulSoup # tool for parsing the html

# For creating the data frame and standard library for organising data
import pandas as pd 

from gensim.parsing.preprocessing import remove_stopwords
from io import StringIO
```

```python
URL = "https://www.presidency.ucsb.edu/documents/inaugural-address-53"
page = requests.get(URL)
page.encoding = 'ISO-885901'
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify()) # to print the source code of webpage for inspection
```

```python
biden = soup.find_all(class_ = "field-docs-content")
for i in biden[:1]:
    print(i)
```

```python
result = remove_stopwords(i.text)
```

```python
result
```

```python
data = StringIO(result)
df = pd.read_csv(data)
```
